```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

SalesDF = pd.read_csv("Sales.csv")

RevenueDF = SalesDF.groupby(["Country", "State"])["Revenue"].aggregate(np.sum).unstack()
RevenueDF.assign(Total=RevenueDF.sum(1)).stack().to_frame("Revenue")

# Learn how to add more columns
```
